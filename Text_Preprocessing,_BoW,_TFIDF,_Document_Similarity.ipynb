# More Advanced preprocessing

- normalisasi: menghindari 2  hal yang maknanya sama. saya dan Saya
- tokenisasi : menghindari token yang tidak diproses dengan baik misalnya `halo,` atau  `ini?`

Lebih jauh lagi akan ada teknik pemrosesan seperti:
- stemming: menghindari kesalahan kesamaan makna karena imbuhan seperti `pukul` dan `memukul`
- lematisasi : mengembalikan kata ke bentuk dasarnya seperti `jum'at` menjadi `jumat`

package yang terkenal untuk stemming: Satrawi, SpaCy
package yang terkenal untuk lematisasi: SpaCy

# Dataset Splitting

```python
X = df.Teks
y = df.label

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

# Training

```python
from sklearn.feature_extraction.text import TfidfVectorizer

## logistic regression
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import RandomizedSearchCV
from jcopml.tuning import random_search_params as rsp
```

```python
pipeline = Pipeline([
    ('prep', TfidfVectorizer(tokenizer=word_tokenize, stop_words=sw_indo)),
    ('algo', LogisticRegression(multi_class='ovr', solver='lbfgs', n_jobs=-1, random_state=42))
])

LogReg = RandomizedSearchCV(pipeline, rsp.logreg_params, cv=3, n_iter=50, n_jobs=-1, verbose=1, random_state=42)
LogReg.fit(X_train, y_train)

print(LogReg.best_params_)
print(LogReg.score(X_train, y_train), LogReg.best_score_, LogReg.score(X_test, y_test))
```

```python
def extract_keywords_tfidf(doc, tfidf, topk=10):
  matrix = tfidf.transform([doc])
  vocab = tfidf.get_feature_names_out()

  sorted_tfidf = matrix[0].toarray()[0].argsort()
  return [vocab[idx] for idx in reversed(sorted_tfidf[-topk:])]
```

# Topic Modelling

```python
vocab = bow.get_feature_names_out()
```

# Latent Semantic Analysis (LSA)

```python
from sklearn.decomposition import TruncatedSVD
```

```python
lsa = TruncatedSVD(n_components=10, n_iter=10, random_state=42)
```

```python
lsa_matrix = lsa.fit_transform(bow_matrix)
```

```python
print(bow_matrix.shape) #hidden
print(lsa_matrix.shape) # weight / code
print(lsa.components_.shape) # fitur / topic
```

```python
def get_topic(model):
  return [[vocab[idx] for idx in reversed(comp.argsort()[-6:]) if vocab[idx].isalnum()]
          for comp in model.components_]
```

```python
get_topic(lsa)
```
